```python
import pandas as pd
```

```python
df = pd.read_excel("dashboard-data-april-download.xlsx", sheet_name="2. Harmonized Indicators")
```

```python
df = df.drop(columns=[
    'footnote',
    'FCS',
    'survey_producer',
    'survey_link'
])
```

```python
df = df.drop(columns=[
    'row_id',
    'code',
    'region_code'
])
```

```python
df.columns 
```

```python
df.isnull().sum()
```

```python
df['month'] = df['month'].fillna(df['month'].median())
df['urban_rural'] = df['urban_rural'].fillna(df['urban_rural'].mode()[0])
df['GDP_pc'] = df['GDP_pc'].fillna(df['GDP_pc'].median())
df['ln_GDP_pc'] = df['ln_GDP_pc'].fillna(df['ln_GDP_pc'].median())
```

```python
df['indicator_topic'] = df['indicator_topic'].fillna("Unknown")
df['indicator_display'] = df['indicator_display'].fillna("Unknown")
df['income_group'] = df['income_group'].fillna("Unknown")
df['unit_measure'] = df['unit_measure'].fillna("Unknown")
df['measure_type'] = df['measure_type'].fillna("Unknown")
df['weight_type'] = df['weight_type'].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['GDP_Category'] = pd.qcut(df['GDP_pc'], q=4, labels=[
    'Low GDP', 'Lower-Mid GDP', 'Upper-Mid GDP', 'High GDP'
])
```

```python
df['Value_Category'] = pd.qcut(df['indicator_val'], q=4, labels=[
    'Low', 'Medium', 'High', 'Very High'
])
```

```python
df[['GDP_pc', 'GDP_Category', 'indicator_val', 'Value_Category']].head()
```

```python
selected_indicators = [
    "Average household size",
    "Male respondents (% of respondents weighted by household weight)",
    "Employed respondents who are wage employee (% of working respondents above 18 years old)",
    "Employed respondents who are self-employed (% of working respondents above 18 years old)",
    "In the last 30 days, went without eating for a whole day due to lack of money (% of HHs)",
    "In the last 30 days, you ran out of food due to a lack of money or other resources? (% of HHs)",
    "Level of education - No education",
    "Level of education - Any Primary"
]
df_filtered = df[df['indicator_display'].isin(selected_indicators)]
df_filtered['indicator_display'].nunique()
```

```python
import hvplot.pandas
import panel as pn
pn.extension("tabulator")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
indicator_options = sorted(df_filtered['indicator_display'].dropna().unique().tolist())
country_options = ['All'] + sorted(df_filtered['country'].dropna().unique().tolist())
region_options = ['All'] + sorted(df_filtered['region'].dropna().unique().tolist())
```

```python
year_select = pn.widgets.RadioButtonGroup(
    name='Year',
    options=sorted(df_filtered['year'].dropna().unique().tolist()),
    button_type='success'
)
sidebar_widgets = pn.Column(
    year_select
)
```

```python
filtered_data = pn.bind(
    lambda year: df_filtered[df_filtered['year'] == year],
    year_select
)
```

```python
top_countries_plot = pn.bind(
    lambda data: (
        data.groupby('country', as_index=False)['indicator_val']
        .mean()
        .sort_values('indicator_val', ascending=False)
        .head(10)
        .hvplot.bar(
            x='country',
            y='indicator_val',
            title='Top 10 Countries',
            color='#f4a261',
            alpha=0.9,
            width=520,
            height=340,
            rot=50,
            grid=True,
            hover=True
        )
        .opts(
            padding=0.1,     
            xlabel='Country',
            ylabel='Indicator Value'
        )
    ) if not data.empty else pn.pane.Markdown("No data available."),
    filtered_data
)
```

```python
distribution_plot = pn.bind(
    lambda data: (
        data[data['indicator_val'] <= data['indicator_val'].quantile(0.95)]
        .hvplot.hist(
            y='indicator_val',
            bins=15,
            title='Distribution of Indicator Values',
            color='#7fb069',
            alpha=0.85,
            width=420,
            height=300,
            grid=True
        )
        .opts(
            xlabel='Indicator Value',
            ylabel='Frequency',
            tools=['hover'],
            toolbar='above',
            fontsize={
                'title': 14,
                'labels': 11,
                'xticks': 9,
                'yticks': 9
            }
        )
    ) if not data.empty else pn.pane.Markdown("No data available."),
    filtered_data
)
```

```python
gdp_scatter = pn.bind(
    lambda data: (
        data[['income_group', 'ln_GDP_pc', 'indicator_val']]
        .dropna()
        .groupby(['income_group', 'ln_GDP_pc'], as_index=False)
        .mean()
        .hvplot.scatter(
            x='ln_GDP_pc',
            y='indicator_val',
            by='income_group',
            title='Relationship between GDP (Log Scale) and Indicator',

            width=800,
            height=320,

            size=55,
            alpha=0.7,

            color_key={
                'Low income': '#e76f51',
                'Lower middle income': '#f4a261',
                'Upper middle income': '#2a9d8f',
                'High income': '#264653'
            },

            legend='right',
            grid=True,
            hover=True,
        )
        .opts(
            xlabel='Log GDP per Capita',
            ylabel='Indicator Value',
            fontsize={
                'title': 14,
                'labels': 11,
                'xticks': 9,
                'yticks': 9
            }
        )
    ) if not data.empty else pn.pane.Markdown("No data available."),
    filtered_data
)
```

```python
all_regions = sorted(df_filtered['region'].dropna().unique().tolist())

region_pie = pn.bind(
    lambda data: (
        pn.pane.Markdown("No data available.")
        if data.empty else
        (
            lambda fig_ax, region_df: (
                fig_ax[1].pie(
                    region_df['indicator_val'],
                    labels=region_df['region'],
                    autopct=lambda p: f'{p:.1f}%' if p > 0 else '',  # hide 0% labels
                    startangle=90,
                    colors=['#264653', '#2a9d8f', '#e9c46a', '#f4a261', '#e76f51', '#8ecae6'],

                    # no white gaps
                    wedgeprops={'edgecolor': 'none'}
                ),

                fig_ax[1].set_title('Regional Distribution', fontsize=15),

                fig_ax[0].subplots_adjust(left=0, right=1, top=0.9, bottom=0),

                plt.close(fig_ax[0]),

                pn.pane.Matplotlib(fig_ax[0], width=500, height=380, tight=True)
            )[-1]
        )(
            plt.subplots(figsize=(6, 6)),

            # 🔥 THIS IS THE IMPORTANT FIX
            data.groupby('region')['indicator_val']
            .mean()
            .reindex(all_regions, fill_value=0)
            .reset_index()
        )
    ),
    filtered_data
)
```

```python
df_corr = df_filtered.copy()

df_corr['region_code'] = df_corr['region'].astype('category').cat.codes
df_corr['income_code'] = df_corr['income_group'].astype('category').cat.codes
df_corr['industry_code'] = df_corr['industry'].astype('category').cat.codes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_heatmap = pn.bind(
    lambda data: (
        pn.pane.Markdown("No data available.")
        if data.empty else
        (
            lambda fig_ax: (
                sns.heatmap(
                    data[['indicator_val', 'GDP_pc', 'ln_GDP_pc', 'sample_total']]
                    .dropna()
                    .corr(),

                    # 🎨 clean color palette
                    cmap='coolwarm',

                    # ✨ readability
                    annot=True,
                    fmt='.2f',

                    # 🔥 clean look
                    linewidths=0.5,
                    square=True,

                    ax=fig_ax[1]
                ),

                fig_ax[1].set_title('Correlation Heatmap', fontsize=15),

                # 🔥 remove extra white space
                fig_ax[0].tight_layout(),

                plt.close(fig_ax[0]),

                # 📏 good size for dashboard
                pn.pane.Matplotlib(fig_ax[0], width=600, height=350, tight=True)
            )[-1]
        )(plt.subplots(figsize=(6, 5)))
    ),
    filtered_data
)
```

```python
dashboard = pn.Row(
    pn.Column(
        pn.pane.Markdown("""
# Socio-Economic Indicators Dashboard

### Insights
- Top countries chart highlights leading countries.
- Distribution shows spread of values.
- Scatter shows GDP vs indicator relationship.
- Pie shows regional distribution.
- Heatmap shows correlation between numeric variables.

### Filter
Use the year buttons below to update all visuals.
        """),
        year_select,
        width=260,
        sizing_mode='fixed'
    ),

    pn.Column(
        pn.Row(top_countries_plot, distribution_plot, sizing_mode="stretch_width"),
        pn.Spacer(height=12),
        gdp_scatter,
        pn.Spacer(height=12),
        pn.Row(region_pie, corr_heatmap, sizing_mode="stretch_width"),
        sizing_mode="stretch_width"
    ),
    sizing_mode="stretch_width"
)
```

```python
pn.serve(dashboard, show=True)
```
